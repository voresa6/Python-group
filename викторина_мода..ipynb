```python
def ask_question(question, options):
    print(question)
    for i, option in enumerate(options, 1):
        print(f"{i}. {option}")
    user_answer = input("Введите номер вашего ответа: ")
    while not user_answer.isdigit() or int(user_answer) < 1 or int(user_answer) > len(options):
        user_answer = input("Пожалуйста, введите корректный номер вашего ответа: ")
    return int(user_answer)

# Код представляет собой программу, которая задает пользователю серию вопросов и позволяет выбрать один из предложенных вариантов ответа. После того, как пользователь ответит на все вопросы, программа подсчитывает количество правильных ответов и выводит итоговый результат.

questions = [
    "Является ли удачным сочетание красный + розовый? (Да/нет)",
    "Правда ли, что первым модельером, выпустившим духи, был Поль Пуаре в 1911 году (Да/нет)",
    "Что такое ахроматические цвета? (Черный, белый серый / все цвета радуги)",
    "Подойдут ли под фигуру «перевернутый треугольник» узкие джинсы? (Да/нет)",
    "Высокие каблуки изначально создавались мужчинами для мужчин (Правда/не правда)",
    "Правда ли, что периодически в моду возвращаются тренды и мода циклична? (Да/нет)",
    "Париж — мировая столица моды (Верно/не верно)",
    "Правда ли, что первая в мире неделя моды прошла во Флоренции? (Да/нет)",
    "Французские дворянки носили прически, вдохновленные видом кораблей (Правда/не правда)",
    "Правда ли, что средний американец ежегодно выбрасывает 80 фунтов одежды? (Да/нет)",
    "Что было самым первым модным аксессуаром? (Часы/браслеты/сумка)",
    "Верно ли, что для фигуры «груша» для баланса нужен объемный верх? (Верно/не верно)",
    "Индустрия моды подвергается критике за пропаганду нереальных стандартов красоты и объективизацию женщин (Верно/не верно)",
    "Валентино владеет пятью мопсами (Правда/не правда)",
    "Louis Vuitton якобы сжигает все свои старые товары, чтобы сохранить полную эксклюзивность бренда (Правда/не правда)",
    "Первый журнал мод появился в 1486 году (Верно/не верно)",
    "До 1800-х годов не было понятия детской одежды. Дети одевались так же, как взрослые (Правда/не правда)",
    "В Риме фиолетовую одежду носили только аристократы и члены королевской семьи (Правда/не правда)",
    "В оригинальной сказке о Золушке фигурируют меховые тапочки вместо хрустальных (Правда/не правда)",
    "До 19 века модные компании использовали кукол вместо моделей (Правда/не правда)"
]

# Функция ask_question(question, options) принимает вопрос и список вариантов ответа, выводит вопрос и варианты ответа, запрашивает у пользователя номер выбранного ответа и проверяет его на корректность.


answers = [1, 2, 1, 2, 2, 1, 1, 2, 2, 2, 3, 1, 1, 2, 2, 1, 2, 2, 1, 2]

score = 0
for question, answer in zip(questions, answers):
    user_choice = ask_question(question, ["Да", "Нет", "Верно", "Не верно", "Черный, белый серый", "Все цвета радуги"])
    if user_choice == answer:
        score += 1

# Далее в коде определены список вопросов questions и список правильных ответов answers. Затем программа проходит по каждому вопросу из списка questions, задавая их пользователю и сравнивая ответы с правильными ответами из списка answers. За каждый правильный ответ пользователю начисляется балл.

print(f"Вы набрали {score} из {len(questions)} возможных баллов.")

# Наконец, программа выводит общее количество набранных баллов из возможного максимального количества баллов.
```
